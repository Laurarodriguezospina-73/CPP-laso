```
import random
import time

def ejecutar_operaciones():
    print("=== ESTADOS DEL PROCESO ===")

    print("\n [NUEVO] - Creando el proceso...")
    time.sleep(0.5)

    print("\n [LISTO] - Proceso listo para ejecutarse")
    print("   Esperando asignación de CPU...")
    time.sleep(0.5)

    print("\n [EJECUTANDO] - Proceso usando la CPU")
    print("   Cargando el proceso...")
    start_time = time.perf_counter()

    n1 = random.randint(1, 20)
    n2 = random.randint(1, 20)

    print(f"\n   El primer número aleatorio es: {n1}")
    print(f"   El segundo número aleatorio es: {n2}")

    peso_resta = random.randint(0, 9)  # kb
    peso_suma = random.randint(0, 9)   # kb

    print(f"   El peso de la resta es: {peso_resta} kb")
    print(f"   El peso de la suma es: {peso_suma} kb")

    # CONDICIÓN: El proceso más liviano (menor peso) se ejecuta primero
    print(f"\n   Buscando el proceso MÁS LIVIANO...")

    if peso_suma < peso_resta:
        print(f" La SUMA es más liviana ({peso_suma} kb) y se ejecuta PRIMERO")
        suma = n1 + n2
        print(f"      La suma es: {suma}")

        print("\n [ESTADO BLOQUEADO] - Esperando operación")
        time.sleep(0.5)
        print(" E/S completada - Volviendo a EJECUTANDO")

        print(f"\n   Ejecutando RESTA después ({peso_resta} kb)")
        resta = n1 - n2
        print(f"      La resta es: {resta}")

    elif peso_resta < peso_suma:
        print(f" La RESTA es más liviana ({peso_resta} kb) y se ejecuta PRIMERO")
        resta = n1 - n2
        print(f"      La resta es: {resta}")

        print("\n [ESTADO BLOQUEADO] - Esperando operación")
        time.sleep(0.5)
        print("  E/S completada - Volviendo a EJECUTANDO")

        print(f"\n   Ejecutando SUMA después ({peso_suma} kb)")
        suma = n1 + n2
        print(f"      La suma es: {suma}")

    else:
        print(f" Los pesos son IGUALES ({peso_suma} kb), ejecutando SUMA primero")
        suma = n1 + n2
        print(f"      La suma es: {suma}")

        print("\n [ESTADO BLOQUEADO] - Esperando operación")
        time.sleep(0.5)
        print(" E/S completada - Volviendo a EJECUTANDO")

        resta = n1 - n2
        print(f"      La resta es: {resta}")

    print("\n [ESTADO TERMINADO] - Proceso finalizado")
    duration = time.perf_counter() - start_time
    print(f"   El proceso total tardó {duration:.6f} segundos.")
    print("   Memoria liberada. Recursos liberados.")

if __name__ == "__main__":
    ejecutar_operaciones()
```
